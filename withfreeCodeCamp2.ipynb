```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# list of every product
sales['Product'].unique()
```

```python
# bar plot showing the 10 most sold product
sales['Product'].value_counts().head(10).plot(kind='bar', 
figsize=(14,6), ylabel='Quantity',rot=45)
```

```python
# bar plot showing the 10 least sold product
sales['Product'].value_counts().tail(10).plot(kind='bar', 
figsize=(14,6), ylabel='Quantity',rot=45)
```

```python
# relation beetwen unit cost and unit price
sales.plot(kind='scatter', x='Unit_Cost', y='Unit_Price', 
figsize=(6,6), xlabel='Unit Cost', ylabel='Unit Price')
```

```python
# relation beetwen order quantity and profit
sales.plot(kind='scatter', x='Order_Quantity', y='Profit', 
figsize=(6,6), xlabel='Order Quantity', ylabel='Profit')
```

```python
# Grouped box plot per country with profit values
# this is shown by pro
sales[['Profit', 'Country']].boxplot(by='Country', figsize=(10,6))
```

```python
# Grouped box plot per costumer age with country
# this is shown by pro
sales[['Customer_Age', 'Country']].boxplot(by='Country', figsize=(10,6))
```

```python
# Add and calculate a new calculated_date column
# use day, month, year to create a date column (YYYY-MM-DD)
sales['Calculated_Date'] = sales[['Year','Month','Day']].apply(lambda x: '{}-{}-{}'.format(x[0], x[1], x[2]), axis=1)
sales['Calculated_Date'].head()
```

```python
# change calculated date column into a datetime object
sales['Calculated_Date']=pd.to_datetime(sales['Calculated_Date'])
sales['Calculated_Date'].head()
```

```python
# how did sales evolve through the years
# show line plot using calculated date
sales['Calculated_Date'].value_counts().plot(kind='line', figsize=(14,6))
```

```python
# increase $50 revenue to every sale
sales['Revenue']+=50
sales['Revenue'].head()
```

```python
# how many orders were made in canada or france
sales.loc[(sales['Country']=='Canada')|
(sales['Country']=='France')].shape[0]
```

```python
# how many bike racks orders were made from Canada
sales.loc[(sales['Country']=='Canada')
&(sales['Sub_Category']=='Bike Racks')].shape[0]
```

```python
# how many orders were made in each region (state) of france
france_states=sales.loc[(sales['Country']=='France'),'State'].value_counts()
france_states
```

```python
# show bar plot with the results
france_states.plot(kind='bar', figsize=(16,6), ylabel='Frequency')
```

```python
# how many sales were made per category
sales['Product_Category'].value_counts()
```

```python
# show a pie plot with the results
sales['Product_Category'].value_counts().plot(kind='pie', figsize=(6,6))
```

```python
# how many orders were made per accessory sub-category
accessories = sales.loc[sales['Product_Category']=='Accessories',
'Sub_Category'].value_counts()
```

```python
# show bar plot with results
accessories.plot(kind='bar', figsize=(14,6))
```

```python
# how many orders were made per bikes sub-category
bikes = sales.loc[sales['Product_Category']=='Bikes',
'Sub_Category'].value_counts()
bikes
```

```python
# show pie plot with the results
bikes.plot(kind='pie', figsize=(6,6))
```

```python
#which gender has the most amount of sales
sales['Customer_Gender'].value_counts()
```

```python
sales['Customer_Gender'].value_counts().plot(kind='bar',rot=True)
```

```python
# how many sales with more than 500 in Revenue were made by men
sales.loc[(sales['Revenue']>500)&(sales['Customer_Gender']=='M')].shape[0]
```

```python
# Get the top-5 sales with the highest revenue
sales.sort_values(['Revenue'],ascending=False).head(5)
```

```python
# Get the sale with the highest revenue
sales.loc[sales['Revenue']==sales['Revenue']].max()
```

```python
#the mean Order_Quantity of orders with more than 10K in revenue?
sales.loc[sales['Revenue']>=10000,'Order_Quantity'].mean()
```

```python
#the mean Order_Quantity of orders with less than 10K in revenue?
sales.loc[sales['Revenue']<10000,'Order_Quantity'].mean()
```

```python
# how many orders were made in May of 2016
sales.loc[(sales['Year']==2016) & (sales['Month']=='May')].shape[0]
```

```python
# How many orders were made between May and July of 2016?
sales.loc[(sales['Year']==2016) & 
(sales['Month'].isin(['May', 'June', 'July']))].shape[0]
```

```python
# Show a grouped box plot per month with the profit values.
profit_2016 = sales.loc[sales['Year'] == 2016, ['Profit', 'Month']]
profit_2016.boxplot(by='Month', figsize=(14,6))
```
